```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/drive/My Drive/Life_Expectancy_Data.csv")
print(df.columns)
print(df[' BMI '].mean())

# sns.pairplot(df)
myBasicCorr = df.corr()
fig, ax = plt.subplots(figsize=(25,25))
sns.heatmap(myBasicCorr, annot=True, ax=ax, linewidths=.5)
print(len(df['Country'].unique()))
```

```
df = pd.read_csv("/content/drive/My Drive/Life_Expectancy_Data.csv")
df = df.fillna(df.mean())
df = df.replace(to_replace ="Developing", value = 0)
df = df.replace(to_replace ="Developed", value = 1)
df.pop('Country')
print(df.columns)

for c in df.columns:
  sns.jointplot(data=df, y="Life expectancy ", x=c)
```

```
sns.lmplot(data=df, x="Life expectancy ", y="Income composition of resources", markers="+")
```
